# Data Preprocessing

## Notebook 3: Preparing the Data for Machine Learning

In this notebook, we prepare the employee attrition dataset for machine learning by cleaning the data, removing unnecessary features, encoding categorical variables, splitting the dataset into training and testing sets, and scaling numerical features where appropriate.

The objective is to transform the raw dataset into a format that machine learning algorithms can understand and learn from.

## 1. Import Libraries

We import the libraries required for data preprocessing, including data manipulation, data splitting, feature encoding, and feature scaling.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

print(pd.__version__)

```

## 2. Load the Dataset

Load the employee attrition dataset into a pandas DataFrame for preprocessing.

```python
df = pd.read_csv("../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv")

df.head()
```

## 3. Remove Unnecessary Columns

Some columns do not contribute to predicting employee attrition because they either contain the same value for every employee or serve only as unique identifiers.

Removing these columns reduces unnecessary information and simplifies the dataset.

```python
df.drop(
    columns=[
        "EmployeeCount",
        "EmployeeNumber",
        "StandardHours"
    ],
    inplace=True
)

df.head()
```

## 4. Separate Features and Target

Separate the dataset into input features (X) and the target variable (y).

The features contain the information used to make predictions, while the target variable represents whether an employee left the company.

```python
X = df.drop("Attrition", axis=1)

y = df["Attrition"]
```

```python
print(X.shape)
print(y.shape)
```

## 5. Encode the Target Variable

Machine learning algorithms require numerical target values.

The Attrition column is converted from text labels ("Yes" and "No") into numerical values.

```python
label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)
```

```python
print(y[:10])
```

## 6. Encode Categorical Features

Categorical features are converted into numerical representations using One-Hot Encoding.

This prevents the model from interpreting categories as having an ordinal relationship.

```python
X = pd.get_dummies(X, drop_first=True, dtype=int)
```
